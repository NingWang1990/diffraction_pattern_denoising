```python
from scipy import ndimage as ndi
import numpy as np
from skimage import io, img_as_float
import matplotlib.pyplot as plt
from scipy.ndimage import gaussian_filter
```

```python
images = []
for i in range(1,100):
    images.append(img_as_float(io.imread('Patterns/1_'+str(i).zfill(3)+'.bmp')))
```

```python
images = np.array(images)
```

```python
shape = images.shape
images = np.reshape(images,(shape[0],-1))
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(0.60).fit(images)
pca.n_components_
```

```python
components = pca.transform(images)
images_new = pca.inverse_transform(components)
```

```python
images_new = np.reshape(images_new, shape)
```

```python
images = np.reshape(images, shape)
```

```python
plt.imshow(images[16])
```

```python
plt.imshow(images_new[16])
```
